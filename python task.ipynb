# Task 1: NumPy Basics

Write a Python script that creates a NumPy array with random integers and performs basic mathematical operations like addition, subtraction, multiplication, and division on it.

```python
import numpy as np

# Generate random array of size 3 with integers between 1 and 10
arr = np.random.randint(1, 11, 3)
print(" Array:", arr)

# Addition
print("Addition:", arr + 5)

# Subtraction
print("Subtraction:", arr - 3)

# Multiplication
print("Multiplication:", arr * 2)

# Division
# Here, we'll use np.where to avoid dividing by zero
print("Division:", np.where(arr != 0, arr / 2, 0))
```

Task 2: Pandas Data Analysis

Using Pandas, load a dataset of your choice (you can find datasets online or use your own). Clean the data and perform basic data analysis tasks like calculating mean, median, and mode for specific columns.

```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Assuming there are missing values (there aren't in this dataset),
# you could fill them using the mean of the respective column as follows:
# df.fillna(df.mean(), inplace=True)

# Calculate mean, median, and mode for the 'sepal_length' column
mean_val = df['sepal_length'].mean()
median_val = df['sepal_length'].median()
mode_val = df['sepal_length'].mode().iloc[0]  # mode() returns a Series, we extract the first value

print("\nAnalysis for 'sepal_length' column:")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
```

Task 3: Data Visualization
Using Matplotlib, create a bar chart or histogram to visualize the distribution of a dataset of your choice. Label the axes and add a title to the chart.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')

# Calculate the mean sepal length for each species
species_means = df.groupby('species')['sepal_length'].mean()

# Create a bar chart to visualize the mean sepal length for each species
plt.figure(figsize=(7, 5))
species_means.plot(kind='bar', color=['Blue', 'black', 'purple'])
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.title('Mean Sepal Length by Species')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Task 4: NumPy Array Manipulation
Create two NumPy arrays and perform operations to concatenate them vertically and horizontally.


```python
import numpy as np

# Create two NumPy arrays
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7, 8, 9], [10, 11, 12]])

print("Array 1:")
print(array1)

print("\nArray 2:")
print(array2)

# Concatenate arrays vertically
vertical_concat = np.vstack((array1, array2))
print("\nVertical Concatenation:")
print(vertical_concat)

# Concatenate arrays horizontally
horizontal_concat = np.hstack((array1, array2))
print("\nHorizontal Concatenation:")
print(horizontal_concat)
```

Task 6: Matplotlib Customization
Create a line plot or scatter plot using Matplotlib and customize it by adding a legend, different marker styles, and changing the line colors.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')

# Create a scatter plot for each species
species_list = df['species'].unique()

# Define different marker styles and colors for each species
markers = ['o', 's', '^']
colors = ['blue', 'red', 'green']

plt.figure(figsize=(10, 6))

for species, marker, color in zip(species_list, markers, colors):
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], marker=marker, color=color, label=species)

# Customize the plot
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

Task 7: NumPy Statistical Analysis
Use NumPy to calculate descriptive statistics like variance, standard deviation, and correlation coefficients for a dataset of your choice (you can find datasets online or use your own).


```python
import numpy as np
import seaborn as sns

# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')

# Extract the sepal_length and sepal_width columns
sepal_length = df['sepal_length'].values
sepal_width = df['sepal_width'].values

# Calculate variance
variance_length = np.var(sepal_length)
variance_width = np.var(sepal_width)

# Calculate standard deviation
std_dev_length = np.std(sepal_length)
std_dev_width = np.std(sepal_width)

# Calculate correlation coefficient between sepal_length and sepal_width
corr_coefficient = np.corrcoef(sepal_length, sepal_width)[0, 1]

# Display results
print(f"Variance of Sepal Length: {variance_length:.2f}")
print(f"Variance of Sepal Width: {variance_width:.2f}")
print(f"\nStandard Deviation of Sepal Length: {std_dev_length:.2f}")
print(f"Standard Deviation of Sepal Width: {std_dev_width:.2f}")
print(f"\nCorrelation Coefficient between Sepal Length and Sepal Width: {corr_coefficient:.2f}")
```

Task 8: Pandas Data Grouping
Given a dataset (you can choose one or use your own), use Pandas to group data by a specific column and calculate summary statistics (e.g., mean, median) for each group.


```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')

# Group data by 'species' column and calculate mean and median
grouped = df.groupby('species').agg({
    'sepal_length': ['mean', 'median'],
    'sepal_width': ['mean', 'median']
})

# Display the grouped data with summary statistics
print(grouped)
```

Task 9: Matplotlib Subplots
Create a Matplotlib figure with multiple subplots to display different aspects of a dataset or related datasets (you can choose the dataset and the type of plots).


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset from Seaborn
data = sns.load_dataset("iris")

# Create a Matplotlib figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 grid of subplots

# Plot 1: Scatter plot of sepal length vs. sepal width
sns.scatterplot(x="sepal_length", y="sepal_width", data=data, ax=axes[0, 0])
axes[0, 0].set_title("Sepal Length vs. Sepal Width")

# Plot 2: Histogram of petal length
sns.histplot(data["petal_length"], ax=axes[0, 1], kde=True, color="purple")
axes[0, 1].set_title("Petal Length Histogram")

# Plot 3: Box plot of petal width by species
sns.boxplot(x="species", y="petal_width", data=data, ax=axes[1, 0])
axes[1, 0].set_title("Petal Width by Species")

# Plot 4: Violin plot of sepal length by species
sns.violinplot(x="species", y="sepal_length", data=data, ax=axes[1, 1])
axes[1, 1].set_title("Sepal Length by Species")

# Adjust layout
plt.tight_layout()

# Show the figure with subplots
plt.show()
```

Task 10: Pandas Data Visualization
Choose a dataset of your choice and create various visualizations (e.g., bar charts, pie charts, scatter plots) using Pandas to represent different aspects of the data, such as trends or distributions.


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn
data = sns.load_dataset("iris")

# Bar chart: Mean sepal length per species
mean_sepal_length_per_species = data.groupby("species")["sepal_length"].mean()
mean_sepal_length_per_species.plot(kind="bar", color="skyblue")
plt.xlabel("Species")
plt.ylabel("Mean Sepal Length")
plt.title("Mean Sepal Length per Species")
plt.show()

# Pie chart: Proportion of instances by species
species_counts = data["species"].value_counts()
species_counts.plot(kind="pie", autopct='%1.1f%%', colors=["lightcoral", "lightgreen", "lightblue"])
plt.title("Proportion of Instances by Species")
plt.axis("equal")  # Equal aspect ratio ensures a circular pie chart
plt.show()

# Scatter plot: Relationship between sepal length and sepal width
data.plot.scatter(x="sepal_length", y="sepal_width", color="purple", marker="o")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs. Sepal Width")
plt.show()

```
